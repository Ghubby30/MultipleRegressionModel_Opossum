```python
#Convert all of the units to inches
#convert mm to inches
c1= op.iloc[:,0].apply(lambda x:x/25.4)
op.iloc[:,0] = c1
c2 = op.iloc[:,1].apply(lambda x:x/25.4)
op.iloc[:,1] = c2
c3 = op.iloc[:,4].apply(lambda x:x/25.4)
op.iloc[:,4] = c3

#convert cmm to inches
c4 = op.iloc[:,2].apply(lambda x:x/2.54)
op.iloc[:,2] = c4
c5 = op.iloc[:,3].apply(lambda x:x/2.54)
op.iloc[:,3] = c5
c6 = op.iloc[:,-2].apply(lambda x:x/2.54)
op.iloc[:,-2] = c6
c7 = op.iloc[:,-1].apply(lambda x:x/2.54)
op.iloc[:,-1] = c7
op
```

```python
#Start building the Machine Learning model on estimating total body length based on head length, chest circumference, skull width
#create another dataframe with these variables
op1 = op.iloc[:,[0,-2,1, 2]]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Setting the values for the independent variables and the dependent variable.
X = op1.iloc[:,[0,1,2]]
Y = op1.iloc[:,3]
```

```python
#Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 100)
```

```python
#import linear model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
#fit the data
lr.fit(X_train,Y_train)
```

```python
#Intercept
c = lr.intercept_
```

```python
#slopes of the variables
m = lr.coef_

```

```python
#All constants in the model
print(c,m)
```

```python
#Test the accuracy of the model with a prediction part 1
Y_pred_train = lr.predict(X_train)
```

```python
#Test the accuracy of the model with a pediction part 2
Y_pred_train
```

```python
#Create a scatterplot to visualize the accuracy of the data Part 1
plt.scatter(Y_train, Y_pred_train)
plt.xlabel("Actual Length in Inches (in.)")
plt.ylabel("Predicted Length in Inches (in.)")
plt.figure(figsize=(12,12))
plt.show()


```

```python
#How good was the maodel statistically ?  Import the R^2 stat.
from sklearn.metrics import r2_score
```

```python
#R^2 stat
r2_score(Y_train,Y_pred_train)
```

```python
#Prediction part 2
Y_pred_test = lr.predict(X_test)
```

```python
#Create a scatterplot to visualize the accuracy of the data Part 2
plt.scatter(Y_test, Y_pred_test)
plt.xlabel("Actual Length in Inches (in.)")
plt.ylabel("Predicted Length in Inches (in.)")
plt.figure(figsize=(12,12))
plt.show()

```

```python
# R stat
r2_score(Y_test,Y_pred_test)
```
